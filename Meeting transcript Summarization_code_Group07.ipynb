Step 2: Text Preprocessing (Cleaning Transcripts for ML Models)

```python
import re
import pandas as pd

# Function to clean transcripts
def preprocess_text(text):
    """
    Cleans text by removing speaker labels, numbers, URLs, and redundant spaces.
    """
    if not isinstance(text, str) or text.strip() == '':
        return "No valid transcript available"

    text = text.lower()  # Convert to lowercase
    text = re.sub(r'speaker \d+:', '', text)  # Remove speaker labels
    text = re.sub(r'\[.*?\]', '', text)  # Remove bracketed content
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'(?<!\S)(\d+)(?!\S)', r'\1', text)  # Retain numbers but remove unnecessary words
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces

    return text

# Load dataset
train_df = pd.read_csv("/content/train_df.csv")

# Check if "Transcript" column exists
if "Transcript" in train_df.columns:
    # Apply preprocessing
    train_df["Processed_Transcript"] = train_df["Transcript"].astype(str).apply(preprocess_text)
    print("\n‚úÖ 'Processed_Transcript' column successfully added!\n")
else:
    print("\nüö® 'Transcript' column not found in train_df!\n")

# Display available columns
print("Available Columns in train_df:", train_df.columns.tolist())
```

Model : Textrank Summarization

```python
import re
import numpy as np
import networkx as nx
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Function to apply improved TextRank summarization
def textrank_summarization(text, num_sentences=4):
    """
    Uses TextRank algorithm with forced priority for numerical values, financial terms,
    and key decisions while filtering out generic meeting phrases.
    """
    sentences = re.split(r'(?<=[.!?]) +', text)

    if len(sentences) <= num_sentences:
        return "\n".join(sentences)

    vectorizer = TfidfVectorizer(stop_words='english')
    sentence_vectors = vectorizer.fit_transform(sentences)

    similarity_matrix = np.dot(sentence_vectors, sentence_vectors.T).toarray()
    graph = nx.from_numpy_array(similarity_matrix)
    scores = nx.pagerank(graph)

    # Convert dictionary to list of tuples (sentence score, sentence text)
    ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)

    # Define filtering rules with dataset-extracted generic phrases
    generic_phrases = {
        "thank you. speaker", "and i think", "my name is", "i want to", "and so i", "thank you very",
        "thank you for", "so i think", "i think that", "this is a", "and i want", "i'm going to",
        "we're going to", "and, you know,", "i just want", "so thank you", "and i know", "and i just",
        "i know that", "we have a", "i think it's", "next speaker, please.", "but i think",
        "i don't know", "so i just", "and this is", "and so we", "thank you so", "i'd like to",
        "and i would", "i think we", "so this is", "one of the", "i think the", "i just wanted",
        "you know, i", "so i want", "president. speaker 0:", "and then the", "i have a",
        "so we have", "are there any", "and thank you", "it has been", "i would like",
        "and i don't", "thank you, madam", "so i would", "i don't think", "we need to",
        "thank you, councilman", "thank you, councilmember", "thank you, councilwoman",
        "today we are discussing", "let‚Äôs begin", "let‚Äôs start with", "moving forward", "we are here to discuss",
        "let‚Äôs move on", "before we continue", "next topic", "as we all know", "we are gathered here",
        "good morning", "good afternoon", "we are discussing", "welcome everyone", "our agenda today",
        "let‚Äôs talk about", "let‚Äôs go over", "thank you for joining", "thank you all for being here"
    }

    question_words = {"should", "could", "can", "what", "why", "how"}

    important_sentences = []

    for score, sentence in ranked_sentences:
        # Skip generic phrases (fixed indentation error)
        if any(re.match(rf"^{re.escape(phrase)}", sentence.lower()) for phrase in generic_phrases):
            continue

        # Skip generic questions like "Should we set a deadline?"
        if sentence.lower().split()[0] in question_words:
            continue

        # Ensure numbers (like 15%, 40%) are retained correctly
        if re.search(r'\b\d+%?\b', sentence.lower()):
            score += 5.0  # Even higher weight for numbers

        if re.search(r'increase|reduce|sales|revenue|performance|target|improve|automation|efficiency|cost|growth|profit|loss|investment|reduction|optimization|budget|forecast|projection', sentence.lower()):
            score += 3.5  # Boost financial and key decision-making terms

        important_sentences.append((score, sentence))

    # Re-rank based on adjusted scores
    final_sentences = sorted(important_sentences, reverse=True)[:num_sentences]

    # Format summary with bullet points and highlights
    summary = "\n".join([f"‚úÖ **{s[1]}**" for s in final_sentences])

    return summary


# ‚úÖ Apply TextRank summarization to dataset
if "Processed_Transcript" in train_df.columns:
    train_df["Generated_Summary_TextRank"] = train_df["Processed_Transcript"].dropna().apply(lambda x: textrank_summarization(x, num_sentences=3))
else:
    print("üö® 'Processed_Transcript' column not found in train_df!")

# ‚úÖ Display results
dataframe=train_df[['Processed_Transcript', 'Generated_Summary_TextRank']].head(5)
dataframe
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Function to apply LSA summarization with dynamic sentence selection
def lsa_summarization(text, num_sentences=3):
    """
    Uses Latent Semantic Analysis (LSA) to extract key sentences.
    Handles short transcripts and avoids out-of-range errors.
    """
    sentences = re.split(r'(?<=[.!?]) +', text)

    # If there are fewer sentences than required, return all available sentences
    if len(sentences) <= num_sentences:
        return " ".join(sentences)

    # Convert sentences into TF-IDF vectors
    vectorizer = TfidfVectorizer(stop_words='english')
    sentence_vectors = vectorizer.fit_transform(sentences)

    # Dynamically set the number of components to avoid errors
    n_components = min(1, sentence_vectors.shape[1])
    svd = TruncatedSVD(n_components=n_components)
    svd.fit(sentence_vectors)

    # Get importance scores and ensure correct indexing
    num_sentences_to_extract = min(num_sentences, len(sentences), len(svd.components_[0]))

    # Fix indexing error by ensuring we do not select more sentences than available
    if num_sentences_to_extract == 0:
        return sentences[0]  # If no components are available, return the first sentence

    importance = np.argsort(svd.components_[0])[-num_sentences_to_extract:]

    # Extract the most relevant sentences safely
    summary = " ".join([sentences[i] for i in sorted(importance) if i < len(sentences)])
    return summary

# Apply the updated LSA summarization
train_df["Generated_Summary_LSA"] = train_df["Processed_Transcript"].apply(lambda x: lsa_summarization(x, num_sentences=3))

# Display a sample result
print(train_df[['Processed_Transcript', 'Generated_Summary_LSA']].head(5))
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD

# Function to apply LSA summarization with improved handling of empty summaries
def improved_lsa_summarization(text, num_sentences=3):
    """
    Uses LSA to extract key sentences, ensuring that it does not return empty summaries.
    """
    sentences = re.split(r'(?<=[.!?]) +', text)

    # If there are fewer sentences than required, return all available sentences
    if len(sentences) <= num_sentences:
        return " ".join(sentences)

    # Convert sentences into TF-IDF vectors
    vectorizer = TfidfVectorizer(stop_words='english')
    sentence_vectors = vectorizer.fit_transform(sentences)

    # Adjust number of components dynamically to avoid empty outputs
    n_components = min(1, sentence_vectors.shape[1])
    svd = TruncatedSVD(n_components=n_components)
    svd.fit(sentence_vectors)

    # Ensure at least one valid sentence is selected
    num_sentences_to_extract = min(num_sentences, len(sentences), len(svd.components_[0]))

    if num_sentences_to_extract == 0:
        return sentences[0] if len(sentences) > 0 else "No summary available"  # Always return at least one sentence


    importance = np.argsort(svd.components_[0])[-num_sentences_to_extract:]

    # Extract the most relevant sentences safely
    summary = " ".join([sentences[i] for i in sorted(importance) if i < len(sentences)])

    # If the summary is empty, use a fallback strategy
    if not summary.strip():
        summary = " ".join(sentences[:2])  # Return first two sentences if LSA fails

    return summary

# Apply the improved LSA summarization function
train_df["Generated_Summary_LSA"] = train_df["Processed_Transcript"].apply(lambda x: improved_lsa_summarization(x, num_sentences=3))

# Display a sample result
print(train_df[['Processed_Transcript', 'Generated_Summary_LSA']].head(5))
```

```python
from nltk.translate.meteor_score import meteor_score

# ‚úÖ Function to Compute METEOR Score Without Using NLTK Tokenizer
def compute_meteor_score(reference_summary, generated_summary):
    """
    Computes METEOR score for reference vs. generated summary.
    This version avoids `punkt` tokenizer issues.
    """
    # Use simple `split()` instead of `nltk.word_tokenize()`
    reference_tokens = reference_summary.lower().split() if isinstance(reference_summary, str) else reference_summary
    generated_tokens = generated_summary.lower().split() if isinstance(generated_summary, str) else generated_summary

    return meteor_score([reference_tokens], generated_tokens)

# ‚úÖ Apply METEOR Score to Dataset
if "Summary" in train_df.columns and "Generated_Summary_TextRank" in train_df.columns:
    train_df["METEOR_Score"] = train_df.apply(
        lambda row: compute_meteor_score(str(row["Summary"]), str(row["Generated_Summary_TextRank"])), axis=1
    )

    print("‚úÖ METEOR scores computed successfully!")

    # ‚úÖ Display Table
train_df[['Summary', 'Generated_Summary_TextRank', 'METEOR_Score']].head(5)

```

```python
import spacy
from rouge_score import rouge_scorer
from nltk.translate.meteor_score import meteor_score

# ‚úÖ Load SpaCy for tokenization
nlp = spacy.load("en_core_web_sm")

# ‚úÖ Function to compute ROUGE scores
def compute_rouge(reference_summary, generated_summary):
    """
    Computes ROUGE scores between the reference and generated summary.
    """
    scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)
    scores = scorer.score(reference_summary, generated_summary)
    return scores

# ‚úÖ Function to compute METEOR score (Without `punkt`)
def compute_meteor(reference_summary, generated_summary):
    """
    Computes METEOR score between reference and generated summary.
    Uses SpaCy for tokenization instead of `punkt`.
    """
    reference_tokens = [token.text for token in nlp(reference_summary)]
    generated_tokens = [token.text for token in nlp(generated_summary)]
    return meteor_score([reference_tokens], generated_tokens)

# ‚úÖ Apply Evaluation Metrics on Validation & Test Datasets
for df, name in [(val_df, "Validation"), (test_df, "Test")]:
    rouge_text_rank = []
    rouge_lsa = []
    meteor_text_rank = []
    meteor_lsa = []

    for i, row in df.iterrows():
        reference_summary = row["Summary"]
        generated_summary_tr = row["Generated_Summary_TextRank"]
        generated_summary_lsa = row["Generated_Summary_LSA"]

        # Compute ROUGE & METEOR Scores
        rouge_text_rank.append(compute_rouge(reference_summary, generated_summary_tr))
        rouge_lsa.append(compute_rouge(reference_summary, generated_summary_lsa))
        meteor_text_rank.append(compute_meteor(reference_summary, generated_summary_tr))
        meteor_lsa.append(compute_meteor(reference_summary, generated_summary_lsa))

    # Store Scores in DataFrame
    df["ROUGE_TextRank"] = rouge_text_rank
    df["ROUGE_LSA"] = rouge_lsa
    df["METEOR_TextRank"] = meteor_text_rank
    df["METEOR_LSA"] = meteor_lsa

    # ‚úÖ Display results
    df[["ROUGE_TextRank", "ROUGE_LSA", "METEOR_TextRank", "METEOR_LSA"]].head(5)
```

- **ROUGE scores** evaluate how closely the generated summaries match reference summaries in terms of overlapping words and phrases.
- **METEOR scores** evaluate semantic similarity, considering synonyms and sentence structure.

### üîç Observations from Evaluation:
- **TextRank** consistently produced higher average **ROUGE** and **METEOR** scores across validation and test datasets compared to LSA.
- In the provided results:
    - **ROUGE-1 and ROUGE-L scores** from TextRank indicate better coverage of key content from reference summaries.
    - **METEOR scores** from TextRank were significantly higher, highlighting better semantic similarity and sentence structure preservation.

### üéØ Conclusion:

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ‚úÖ Function to Compute Text Similarity
def compute_text_overlap(text1, text2):
    vectorizer = CountVectorizer().fit_transform([text1, text2])
    vectors = vectorizer.toarray()
    return cosine_similarity([vectors[0]], [vectors[1]])[0][0]

# ‚úÖ Compute Overlap Between Generated & Reference Summaries
train_df["TextRank_Overlap"] = train_df.apply(lambda row: compute_text_overlap(row["Generated_Summary_TextRank"], row["Summary"]), axis=1)
train_df["LSA_Overlap"] = train_df.apply(lambda row: compute_text_overlap(row["Generated_Summary_LSA"], row["Summary"]), axis=1)

# ‚úÖ Boxplot for Overlap Comparison
plt.figure(figsize=(10, 5))
sns.boxplot(data=train_df[["TextRank_Overlap", "LSA_Overlap"]])
plt.title("Text Overlap Between Generated & Reference Summaries")
plt.ylabel("Cosine Similarity Score")
plt.show()
```

```python
import gradio as gr
import time
import re
import numpy as np
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob
import matplotlib.pyplot as plt

# Text Preprocessing Function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'speaker \d+:', '', text)  # Remove speaker labels
    text = re.sub(r'\[.*?\]', '', text)  # Remove bracketed content
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Improved TextRank with Proper Formatting for Detailed Summary
def textrank_summarization(text, detail_level):
    sentences = re.split(r'(?<=[.!?]) +', text)
    if len(sentences) <= 4:
        return "\n".join([f"‚úÖ {s}" for s in sentences])

    vectorizer = TfidfVectorizer(stop_words='english')
    sentence_vectors = vectorizer.fit_transform(sentences)
    similarity_matrix = np.dot(sentence_vectors, sentence_vectors.T).toarray()
    graph = nx.from_numpy_array(similarity_matrix)
    scores = nx.pagerank(graph)
    ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)

    num_sentences = 4 if detail_level == "Concise" else max(6, min(10, len(sentences) // 3))  # Dynamic selection
    important_sentences = []
    for score, sentence in ranked_sentences:
        if re.search(r'\b(approve|increase|decrease|report|budget|plan|proposal|allocate|invest|impact|prioritize|delay|hiring|action|schedule|review|upcoming)\b', sentence.lower()):
            score += 3.5  # Boost key decision-making & upcoming plans
        if re.search(r'\b\d+%?\b', sentence):  # Boost numerical data
            score += 4.0
        important_sentences.append(sentence)

    final_sentences = sorted(important_sentences, reverse=True)[:num_sentences]

    return "\n".join([f"‚úÖ {s}" for s in final_sentences])

def analyze_sentiment(text):
    sentiment_score = TextBlob(text).sentiment.polarity

    # Expanded and weighted keywords
    negative_keywords = [
        "decline", "drop", "loss", "decrease", "budget overrun", "delayed", "issues",
        "problems", "risk", "cutbacks", "failed", "layoffs", "underperformance",
        "market downturn", "shortfall", "reduction", "crisis", "turnover",
        "complaints", "behind schedule", "technical challenges", "disruptions",
        "supplier costs", "logistics expenses", "exit interviews", "concerns"
    ]

    positive_keywords = [
        "growth", "improve", "increase", "expand", "success", "gain",
        "higher revenue", "profit", "optimization", "achievement", "efficiency"
    ]

    # Boost weights for keyword occurrences
    negative_count = sum(text.lower().count(word) for word in negative_keywords)
    positive_count = sum(text.lower().count(word) for word in positive_keywords)

    sentiment_score -= negative_count * 0.2  # stronger penalty
    sentiment_score += positive_count * 0.1  # mild positive adjustment

    if sentiment_score > 0.2:
        return "üòä Positive Meeting"
    elif sentiment_score < -0.05:  # Adjusted threshold for negativity
        return "üòü Negative Meeting"
    else:
        return "üòê Neutral Meeting"



# Summary Generator
def generate_summary(transcript, detail_level):
    cleaned_text = preprocess_text(transcript)
    time.sleep(2)  # Simulating processing time
    return textrank_summarization(cleaned_text, detail_level)

# Meeting Insights (Sentiment Analysis + Word Count)
def meeting_insights(transcript):
    if not transcript:
        return "‚ö†Ô∏è No transcript provided", "‚ö†Ô∏è No data to analyze"

    cleaned_text = preprocess_text(transcript)
    sentiment = analyze_sentiment(cleaned_text)
    word_count = len(cleaned_text.split())
    return sentiment, f"Word Count: {word_count} words"

# UI Design with Gradio Blocks
with gr.Blocks(css="""
    body { background-color: #121212; color: #fff; font-family: 'Arial', sans-serif; }
    .gradio-container { max-width: 800px; margin: auto; padding: 20px; }
    .gr-textbox textarea { background-color: #222; color: white; border: 1px solid #555; }
    .gr-button { background-color: #ff9800; color: white; font-weight: bold; }
    .gr-markdown { background-color: #222; padding: 10px; border-radius: 8px; }
""") as demo:
    gr.Markdown("""# üìã AI-Powered Meeting Minutes Generator""")
    gr.Markdown("""### Paste a transcript to generate a summary.""")

    with gr.Tab("üîπ Meeting Summarizer"):
        input_text = gr.Textbox(lines=10, placeholder="Paste your meeting transcript here...")
        detail_level = gr.Radio(["Concise", "Detailed"], label="Summary Type")
        submit_btn = gr.Button("‚ú® Generate Summary")
        output_text = gr.Markdown(label="üìå **Generated Meeting Minutes**")
        submit_btn.click(generate_summary, inputs=[input_text, detail_level], outputs=output_text)

    with gr.Tab("üìä Meeting Insights"):
        insights_input = gr.Textbox(lines=10, placeholder="Paste your transcript to analyze meeting insights")
        analyze_btn = gr.Button("üìà Analyze Meeting")
        sentiment_output = gr.Markdown(label="üîç Meeting Sentiment")
        word_count_output = gr.Markdown(label="üìù Word Count")
        analyze_btn.click(meeting_insights, inputs=insights_input, outputs=[sentiment_output, word_count_output])

demo.launch(share=True)
```
